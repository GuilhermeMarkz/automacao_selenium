```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# abrir o navegador
navegador = webdriver.Chrome()

# acessar o site
navegador.get("http://localhost/Modelo_do_professor/")

# colocar o navegador em tela cheia
navegador.maximize_window()

# Esperar até que o campo de nome esteja visível
WebDriverWait(navegador, 10).until(
    EC.visibility_of_element_located((By.ID, "name"))
)

# escrever no campo de nome
nome_input = navegador.find_element(By.ID, "name")
nome_input.send_keys("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa")

# Esperar até que o campo de email esteja visível
WebDriverWait(navegador, 10).until(
    EC.visibility_of_element_located((By.ID, "email"))
)

# escrever no campo de email
email_input = navegador.find_element(By.ID, "email")
email_input.send_keys("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa")

# Esperar até que o campo de mensagem esteja visível
WebDriverWait(navegador, 10).until(
    EC.visibility_of_element_located((By.ID, "message"))
)

# escrever no campo de mensagem
mensagem_input = navegador.find_element(By.ID, "message")
mensagem_input.send_keys("Ç!@#$%¨&*()_+")

# Esperar até que o botão de envio esteja clicável
WebDriverWait(navegador, 10).until(
    EC.element_to_be_clickable((By.NAME, "submit"))
)

# selecionar e clicar no botão de envio (ou login)
botao_enviar = navegador.find_element(By.NAME, "submit")
botao_enviar.click()

# opcional: esperar para observar o resultado antes de fechar
time.sleep(5)
```
